```python
#Ticketå±žæ€§
import re
def getTicketPrefix(ticket):
    match = re.compile("([a-zA-Z\.\/]+)").search(ticket)
    if match:
        return match.group()
    else:
        return 'U'
def getTicketNumber(ticket):
    match = re.compile("([0-9]+$)").search(ticket)
    if match:
        return  match.group()
    else:
        return '0'

df = pd.read_csv('train.csv')
def processTicket(df):
    #å…ˆåŒ–ä½œå­—ç¬¦ä¸²æ•°ç»„
    df['TicketPrefix'] = df['Ticket'].map(lambda x: getTicketPrefix(x.upper()))
    df['TicketPrefix'] = df['TicketPrefix'].map(lambda x: re.sub('[\.?\/?]', '', x))  # åŽ»æŽ‰./
    df['TicketPrefix'] = df['TicketPrefix'].map(lambda x: re.sub('STON', 'SOTON', x))
    df['TicketPrefix'] = pd.factorize(df['TicketPrefix'])[0]
    df['TicketNumber'] = df['Ticket'].map(lambda x: getTicketNumber(x))
    df['TicketNumberLength'] = df['TicketNumber'].map(lambda x: len(x)). astype(int)
    df['TicketNumberStart'] = df['TicketNumber'].map(lambda x: x[0:1]).astype(int)
    df['TicketNumber'] = df['TicketNumber'].astype(int)
    return df
df = processTicket(df)
df
```

```python
#-*-coding:utf-8-*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import  preprocessing
from sklearn.ensemble import RandomForestRegressor
#ä¸€ä¸ªå±žæ€§ä¸€ä¸ªå±žæ€§çš„æŽ¢ç´¢ä»¥åŠå¤„ç†
#é€šè¿‡exploreæŽ¢æµ‹åˆ°å…·ä½“çš„æ•°å€¼å±žæ€§ä»¥åŠç¼ºçœå€¼å±žæ€§ä»¥åŠå­—ç¬¦å±žæ€§
#é¢˜æ„ä¸Šç»™å‡ºäº†such as women, children, and the upper-class ä¹Ÿå°±æ˜¯ç­‰çº§ï¼Œå¹´é¾„ï¼Œæ€§åˆ«éžå¸¸ç›¸å…³
#åŒæ—¶æŽ¢ç´¢åˆ°stdå˜æ¢è¾ƒå¤§çš„æ˜¯ageå’ŒFareå±žæ€§

#Pclass
#exploreåŽå¯èƒ½ç­‰çº§è¶Šé«˜èŽ·æ•‘çš„å¯èƒ½æ€§è¶Šå¤§
#ä½†æ˜¯ä¸èƒ½ç¡®å®šæ˜¯å¦ä¸€å®šæ‰€ä»¥æˆ‘ä»¬ç”¨è™šæ‹ŸåŒ–æ¥å¤„ç†one-hot_encoderç›¸å½“äºŽ

def processPclass(df):
    df.Pclass[df.Pclass.isnull()] = df.Pclass.dropna().mode().values
    dummies_df = pd.get_dummies(df['Pclass'],prefix='Pclass')#å‰ç¼€
    df = pd.concat([df,dummies_df],axis=1)
    return df

#Nameå±žæ€§
#æåˆ°çš„å…³é”®å±žæ€§æœ‰å¹´é¾„æ€§åˆ«ï¼Œç­‰çº§ å‡æœ‰è¯´æ˜Žæ•´ä¸ªå±žæ€§æ¯”è¾ƒé‡è¦
#é•¿åº¦é•¿è¯´æ˜Žè¿™ä¸ªäººå¯èƒ½åœ°ä½é«˜
#å­—ç¬¦ç±»åž‹å¯èƒ½éœ€è¦ç¦»æ•£åŒ–æˆ–è€…è™šæ‹ŸåŒ–
#ç”±äºŽå±žæ€§éœ€è¦å–å€¼ç§ç±»å¤ªå¤šã€‚æˆ‘ä»¬å°†æ‰€æœ‰æ›¿æ¢ä¸ºå‡ ä¸ªå…³é”®çš„å±žæ€§

def substrings_in_string(big_string, substrings):
    for substring in substrings:
        if big_string.find(substring)!=-1:
            return substring
    return np.nan

#æ›¿æ¢è§„åˆ™
def replace_titles(x):
    title = x['Title']
    if title in ['Mr', 'Don', 'Major', 'Capt', 'Jonkheer', 'Rev', 'Col']:
        return 'Mr'
    elif title in ['Master']:
        return 'Master'
    elif title in ['Countess', 'Mme', 'Mrs']:
        return 'Mrs'
    elif title in ['Mlle', 'Ms', 'Miss']:
        return 'Miss'
    elif title == 'Dr':
        if x['Sex'] == 'male':
            return 'Mr'
        else:
            return 'Mrs'
    elif title == '':
        if x['Sex'] == 'male':
            return 'Master'
        else:
            return 'Miss'
    else:
        return title
from  sklearn  import preprocessing
le = preprocessing.LabelEncoder()
def processName(df):
    # title_list = ['Mrs', 'Mr', 'Master', 'Miss', 'Major', 'Rev',
    #               'Dr', 'Ms', 'Mlle', 'Col', 'Capt', 'Mme', 'Countess',
    #               'Don', 'Jonkheer']
    # df['Title'] = df['Name'].map(lambda x:substrings_in_string(str(x),title_list))
    # #ä»ç„¶å¤ªå¤šç±»åž‹ï¼Œæˆ‘ä»¬ç®€åŒ–å°±4ç§ Mr Mrs Miss Master
    # df['Title'] = df.apply(replace_titles,axis=1)
    # #å­—ç¬¦ç±»åž‹ç¦»æ•£åŒ–
    # le.fit(df['Title'])
    # title_le = le.transform(df['Title'])
    # df['Title_id'] = title_le.astype(np.int32)
    df['Names'] = df['Name'].map(lambda x: len(re.split(' ', x)))  # è¿™ä¸ªåå­—æœ‰å‡ ä¸ªè¯ä¹Ÿå°±æ˜¯é•¿åº¦
    df['Title'] = df['Name'].map(lambda x: re.compile(",(.*?)\.").findall(x)[0])#å…¶ä¸­ç¬¬ä¸€ä¸ª.æ˜¯ä»»æ„å­—ç¬¦
    df['Title'] = df.apply(replace_titles,axis=1)
    # #å­—ç¬¦ç±»åž‹ç¦»æ•£åŒ–
    le.fit(df['Title'])
    title_le = le.transform(df['Title'])
    df['Title_id'] = title_le.astype(np.int32)
    return df


#Sexå±žæ€§
#æŽ¢ç´¢åŽå‘çŽ°å¥³æ€§å­˜æ´»çŽ‡é«˜å¤§æ¦‚ç™¾åˆ†ä¹‹75å·¦å³æ‰€ä»¥æ˜¯å…³é”®å±žæ€§
#æ–‡æœ¬-ã€‹æ•°å€¼ ç”±äºŽç”·å¥³æœ‰åŒºåˆ«æ‰€æœ‰è½¬æ¢æˆæ•°å€¼å°±è¡Œäº†
def processSex(df):
    df['Gender'] = df['Sex'].map({'female':1,'male':2}).astype(np.int32)
    return df


#Ageå±žæ€§
#stdå˜æ¢è¾ƒå¤§ã€‚ç¼ºçœå€¼æœ‰
#20-30æœ€å¤šäººç„¶åŽå°å­©ç¬¬äºŒå¤šã€‚æœ€å°‘çš„æ˜¯è€äºº
#å°å­©èŽ·æ•‘ä»¥åŠè€äººèŽ·æ•‘å‡ çŽ‡è¾ƒé«˜ã€‚å…¶ä»–å·®ä¸å¤šä¸€èˆ¬å·¦å³

def setMissingData(df,features=[],missFeature='Age'):
    feature_df = df[features]
    X = feature_df[df[missFeature].notnull()].as_matrix()[:,1::]
    y = feature_df[df[missFeature].notnull()].as_matrix()[:,0]
    rtr = RandomForestRegressor(n_estimators=2000,n_jobs=-1)#æ— é™åˆ¶å¤„ç†æœº
    rtr.fit(X,y)
    predicitedAges = rtr.predict(feature_df[df[missFeature].isnull()].as_matrix()[:,1:])
    df.loc[(df[missFeature].isnull()),missFeature] = predicitedAges
    return df

# def setMissingAges(df):
#
#     age_df = df[['Age', 'Embarked', 'Fare', 'Parch', 'SibSp', 'Title_id', 'Pclass', 'Names', 'CabinLetter']]
#     X = age_df.loc[(df.Age.notnull())].values[:, 1::]
#     y = age_df.loc[(df.Age.notnull())].values[:, 0]
#
#     rtr = RandomForestRegressor(n_estimators=2000, n_jobs=-1)
#     rtr.fit(X, y)
#
#     predictedAges = rtr.predict(age_df.loc[(df.Age.isnull())].values[:, 1::])
#     df.loc[(df.Age.isnull()), 'Age'] = predictedAges
#     return  df
def processAge(df):
    #å…ˆå¡«ç¼ºçœå€¼
    #é¢„æµ‹çš„æ–¹æ³•RandomForest
    df = setMissingData(df, features=['Age','Embarked','Fare', 'Parch', 'SibSp', 'Title_id','Pclass','Names','CabinLetter'], missFeature='Age')
    #df = setMissingAges(df)
    #æ­¤å¤„ç”¨ä¸­ä½æ•°ä»¥åŠå‡å€¼å¡«å……ä½†æ˜¯éœ€è¦å…ˆåˆ†å±‚å†æ±‚å‡å€¼ã€‚
    # mean_master = np.average(df['Age'][df.Title=='Master'].dropna())
    # mean_mr = np.average(df['Age'][df.Title=='Mr'].dropna())
    # mean_miss = np.average(df['Age'][df.Title=='Miss'].dropna())
    # mean_mrs = np.average(df['Age'][df.Title=='Mrs'].dropna())
    # df.loc[(df.Age.isnull())&(df.Title=='Master'),'Age'] = mean_master
    # df.loc[(df.Age.isnull()) & (df.Title == 'Mr'), 'Age'] = mean_mr
    # df.loc[(df.Age.isnull())&(df.Title=='Miss'),'Age'] = mean_miss
    # df.loc[(df.Age.isnull()) & (df.Title == 'Mrs'), 'Age'] = mean_mrs
    scaler = preprocessing.StandardScaler()
    df['Age_scaled'] = scaler.fit_transform(df['Age'])
    #ç‰¹åˆ«æåˆ°è€äººå°å­©ã€‚é‚£ä¹ˆæ˜¾ç„¶è¦ç¦»æ•£åŒ–å¹´é¾„
    # bin into quartiles and create binary features
    #æŒ‰ç…§é¢‘çŽ‡æŽ¥è¿‘çš„ç±»åˆ«ç¼–å·åœ¨ä¸€èµ·
    df['Age_bin'] = pd.qcut(df['Age'],4)
    #æ­¤æ—¶stdæ¯”è¾ƒå°äº†å¯ä»¥ä¸å˜æ¢
    df['Age_bin_id'] = pd.factorize(df['Age_bin'])[0]+1

    df['Child'] = (df['Age']<13).astype(int)

    #å˜åŒ–ä¸å¤§
    # from sklearn import  preprocessing
    # scaler = preprocessing.StandardScaler()
    # df['Age_bin_id_scaled'] = scaler.fit_transform(df['Age_bin_id'])
    return  df


#å¤„ç†å…„å¼Ÿå§å¦¹å’Œé…å¶çš„æ•°é‡
#è”æƒ³åˆ°å¦‚æžœè¿™ä¸ªæ•°ç›®è¾ƒå¤šçš„è¯å¯èƒ½æ˜¯å…„å¼Ÿçœ‹å¹´é¾„è¾ƒå°çš„è¯å°±æ˜¯å­©å­
#è€Œè¿™ä¸¤ä¸ªå±žæ€§æˆ‘ä»¬è”æƒ³åˆ°å…¶å®žå¯ä»¥èžåˆæˆä¸€ä¸ªå®¶åº­å±žæ€§ã€‚è€Œä¸”äººæ•°è¶Šå¤šæƒ…å†µæœ‰å¯èƒ½æ²¡èŽ·æ•‘
#0æ¯”è¾ƒå¤šæˆ‘ä»¬ç®—ä¸Šè‡ªå·±
def processSibsp(df):
    df['SibSp'] = df['SibSp'] + 1
    return df

def processParch(df):
    df['Parch'] = df['Parch'] + 1
    return df

def processFamily(df):
    df = processSibsp(df)
    df = processParch(df)
    df['Family'] = df['SibSp'] + df['Parch']
    return df

#å¤„ç†Fare stdå˜æ¢è¾ƒå¤§Scaledå¿…é¡»çš„ã€‚åŒæ—¶0æ¯”è¾ƒå¤šå¹³æ»‘åŒ–.è€Œå’Œå…³é”®å±žæ€§çŒœæƒ³å’Œupper classæœ‰å…³
#æ´»ä¸‹æ¥çš„äººï¼ŒFareæ€»èŠ±è´¹è¾ƒå¤š
#ç®—ä¸€ä¸‹æ¯ä¸ªäººçš„èŠ±è´¹
# çš„å¯èƒ½æ€§æ¯”è¾ƒå¤§å¯èƒ½äººç‰©è¶Šæœ‰
def processFare(df):
    #stdå¤„ç†æ–¹å¼æœ‰å¾ˆå¤šã€‚æœ€åŸºæœ¬æ˜¯scaled.è¿˜å¯ä»¥ç¦»æ•£åŒ–
    #å…ˆå¹³æ»‘åŒ–ä½¿å¾—0æ²¡æœ‰é‚£ä¹ˆå¤š
    df.loc[(df.Fare==0),'Fare'] = df['Fare'][df.Fare.nonzero()[0]].min()/100
    sclar = preprocessing.StandardScaler()
    df['Fare_scaled'] = sclar.fit_transform(df['Fare'])
    df['Fare_bin'] = pd.qcut(df['Fare'],4)
    df['Fare_bin_id'] = pd.factorize(df['Fare_bin'])[0] + 1
    return df

#æˆ‘ä»¬çŒœæµ‹å’Œç­‰çº§æœ‰å…³ã€‚ç±»ä¼¼åå­—å¤„ç†æ–¹å¼æ— æ³•ç¦»æ•£åŒ–ã€‚åªæœ‰ç”¨å‡ ä¸ªä»£è¡¨çš„ç±»åž‹æ¥è¡¨ç¤º
#æˆ‘ä»¬éœ€è¦åˆ†æžç¥¨çš„å‰ç¼€åŒæ—¶å‘çŽ°å’ŒPclassçš„å…³ç³»ä¹Ÿå°±æ˜¯è¯´å‰ç¼€ç¡®å®žæœ‰ç­‰çº§çš„å…³ç³»
#è¿™ç§å­—ç¬¦ä¸²æˆ‘ä»¬éƒ½éœ€è¦å‰–æžå‰ç¼€å’ŒåŽç¼€ç„¶åŽç¼–ç 
#èŽ·å–å‰ç¼€
import re
def getTicketPrefix(ticket):
    match = re.compile("([a-zA-Z\.\/]+)").search(ticket)
    if match:#æœ‰å‰ç¼€
        return match.group()
    else:#æ²¡æœ‰å‰ç¼€
        return 'U'

#èŽ·å–ç¥¨çš„æ•°å­—
def getTicketNumber(ticket):
    match = re.compile("([0-9]+$)").search(ticket)
    if match:
        return  match.group()
    else:
        return '0'

def processTicket(df):
    #å…ˆåŒ–ä½œå­—ç¬¦ä¸²æ•°ç»„
    #strTick = df.Ticket.map(lambda x: str(x))
    #lenTick = df.Ticket.map(lambda x: len(x))
    df['TicketPrefix'] = df['Ticket'].map(lambda x: getTicketPrefix(x.upper()))
    df['TicketPrefix'] = df['TicketPrefix'].map(lambda x: re.sub('[\.?\/?]', '', x))  # åŽ»æŽ‰./
    df['TicketPrefix'] = df['TicketPrefix'].map(lambda x: re.sub('STON', 'SOTON', x))
    df['TicketPrefix'] = pd.factorize(df['TicketPrefix'])[0]
    df['TicketNumber'] = df['Ticket'].map(lambda x: getTicketNumber(x))
    df['TicketNumberLength'] = df['TicketNumber'].map(lambda x: len(x)). astype(int)
    df['TicketNumberStart'] = df['TicketNumber'].map(lambda x: x[0:1]).astype(int)
    df['TicketNumber'] = df['TicketNumber'].astype(int)
    return df


#Cabinæœ‰å¾ˆå¤šç¼ºçœå€¼
#å‘çŽ°æ˜¯å¦ç¼ºçœå¯¹ç»“æžœå½±å“å¾ˆå¤§
#åŒæ—¶å‚æ•°è´¡çŒ®åº¦ä¸å¤Ÿçš„è¯´æ˜Žåˆ’åˆ†ä¸å¤Ÿç»†
#èƒ½ä»£è¡¨ç€ä¸€äº›éšå«ä¿¡æ¯ã€‚æ¯”å¦‚èˆ¹èˆ±å·Cabinè¿™ä¸€å±žæ€§ï¼Œç¼ºå¤±å¯èƒ½ä»£è¡¨å¹¶æ²¡æœ‰èˆ¹èˆ±
#å‘çŽ°å­—æ¯è·Ÿç­‰çº§æœ‰å…³
def getCabinLetter(cabin):
    match = re.compile("([a-zA-Z]+)").search(cabin)
    if match:
        return  match.group()
    else:
        return 'U'

def getCabinNumber(cabin):
    match = re.compile("([0-9]+)").search(cabin)
    if match:
        return match.group()
    else:
        return 0

def processCabin(df):
    df['Cabin'][df.Cabin.isnull()] = 'U0'
    df['CabinLetter'] = df['Cabin'].map(lambda x:getCabinLetter(x))
    df['CabinLetter'] = pd.factorize(df['CabinLetter'])[0]
    df['CabinNumber'] = df['Cabin'].map(lambda x:getCabinNumber(x)).astype(int) + 1 #0å¤ªå¤š
    #stdæ¯”è¾ƒå¤§æ‰€ä»¥æˆ‘ä»¬è¦æ ‡å‡†åŒ–
    sclar = preprocessing.StandardScaler()
    df['CabinNumber'] = sclar.fit_transform(df['CabinNumber'])
    return df



#å¤„ç†Embarked ä¸Šèˆ¹ä½ç½®
#çŒœæƒ³ä¸Šèˆ¹çš„ä½ç½®é è¿‘æ•‘ç”Ÿå¤„ä»¥åŠæ˜¯ä»¥æœ€è¿‘ç­‰çº§é«˜çš„åœ°æ–¹æ‰€ä»¥è¿™ä¸ªå±žæ€§éœ€è¦
#Sçª—å£æœ€å¤šã€‚CèŽ·æ•‘å‡ çŽ‡æœ€å¤§
def processEmbarked(df):
    df.Embarked[df.Embarked.isnull()] = df.Embarked.dropna().mode()#çºµæ•°
    # #å¦‚æžœæ ¹æ®æ•°å­—å˜åŒ–æœ‰ä¸€å®šè¶‹åŠ¿ã€‚
    # df.Embarked = pd.factorize(df['Embarked'])[0]#æ•°å€¼åŒ–
    # #ä¹Ÿå¯ä»¥è™šæ‹ŸåŒ–
    df = pd.concat([df,pd.get_dummies(df['Embarked'],prefix='Embarked')],axis=1)
    df.Embarked = pd.factorize(df['Embarked'])[0]
    return df



#è€ƒè™‘ç»„åˆç‰¹å¾
#è€Œæˆ‘ä»¬ç»„åˆç‰¹å¾å…¶å®žæ˜¯æŠŠæ¯ä¸€ç»´ç‰¹å¾æƒé‡åº”è¯¥ç›¸ç­‰çœ‹å¾…ã€‚æ‰€ä»¥æœ€å¥½éƒ½scaled
def combineFeature(df):
    print("Starting With",df.columns.size,"æ‰‹åŠ¨ç”Ÿæˆç»„åˆç‰¹å¾",df.columns.values)
    numerics = df.loc[:,['Age_scaled']]
    return df


#ç‰¹å¾æå–é™ä½Žç»´



#æ•°æ®æ¸…ç†ä»¥åŠç‰¹å¾å·¥ç¨‹
def clean_and_feature_engineer(df):
    return df




#ä¸»å‡½æ•°
def dataprocess(filename='train.csv'):
    data_train = pd.read_csv(filename)
    data_train = clean_and_feature_engineer(data_train)
    return data_train

```

```python
def processDrops(df):
    #åˆ é™¤å­—ç¬¦å±žæ€§çš„å€¼
    stringsDropList = ['Title', 'Name', 'Cabin', 'Ticket', 'Sex', 'Ticket', 'TicketNumber','Age_bin','Fare_bin']
    df.drop(stringsDropList,axis=1,inplace=True)

    #å‘çŽ°corrç›¸å½“çš„å°
    df_corr = df.drop(['Survived', 'PassengerId'], axis=1).corr(method='spearman')
    mask = np.ones(df_corr.columns.size) - np.eye(df_corr.columns.size)#é»˜è®¤ä¸­é—´ä¸º0ç›¸å½“äºŽè·Ÿè‡ªå·±æ— å…³
    df_corr = mask * df_corr
    #å¾—åˆ°æ•´ä½“çš„åæ–¹å·®çŸ©é˜µ

    #èµ›é€‰å‡ºç›¸å…³æ€§å¤ªé«˜çš„å±žæ€§å†—æ‚å±žæ€§
    drops = []
    for col in df_corr.columns.values:
        if np.in1d([col],drops):#å·²ç»æœ‰äº†
            continue
        corr = df_corr[abs(df_corr[col]) > 0.98].index#ç¬¬colåˆ—ä¸­ç›¸å…³æ€§å¤ªå¤§çš„èˆå¼ƒ
        drops = np.union1d(drops, corr)
    print("\nDropping", drops.shape[0], "highly correlated features...\n")# , drops
    df.drop(drops,axis=1,inplace=True)
    return  df

#æˆ‘ä»¬å¾—åˆ°äº†å…·æœ‰å¤§é‡ç‰¹å¾çš„ç»´åº¦å¾ˆé«˜çš„æ•°æ®é›†ï¼Œç‰¹å¾è¾ƒå¤šä¸èƒ½ç›´æŽ¥ç”¨æ¥ä½œä¸ºæ¨¡åž‹è¾“å…¥ï¼Œä¸€æ˜¯å› ä¸ºè¿™äº›ç‰¹å¾é—´å…·æœ‰å¤šé‡å…±çº¿æ€§ï¼Œ
# å¯èƒ½ ä¼šå¯¼è‡´ç©ºé—´çš„ä¸ç¨³å®šï¼›äºŒæ˜¯å› ä¸ºé«˜ç»´ç©ºé—´æœ¬èº«å…·æœ‰ç¨€ç–æ€§ï¼Œä¸€ç»´æ­£æ€åˆ†å¸ƒæœ‰68%çš„å€¼è½äºŽæ­£è´Ÿæ ‡å‡†å·®ä¹‹é—´ï¼Œè€Œåœ¨åç»´ç©ºé—´ä¸Šåªæœ‰0.02%ï¼›ä¸‰æ˜¯ç”±äºŽè¿‡å¤šçš„å±žæ€§
# ä¼šä½¿æŒ–æŽ˜éœ€è¦å¾ˆé•¿æ—¶é—´ã€‚å¯¹äºŽä¸€äº›æ¨¡åž‹æ¥è¯´ï¼Œæ¯”å¦‚ä½¿ç”¨L1ï¼ˆLassoï¼‰ï¼Œå½“æœ‰å¤§é‡å±žæ€§æ—¶æ•ˆæžœå¾ˆå¥½ï¼Œå› ä¸ºå®ƒå¯ä»¥æœ‰æ•ˆå¿½ç•¥æŽ‰å™ªå£°å˜é‡ã€‚è€Œä¸€äº›æ¨¡åž‹åˆ™å®¹æ˜“è¿‡æ‹Ÿ åˆã€‚
#ä¼šä½¿ç”¨èšç±»åˆ†æžå’Œé™ç»´æ¥å¤„ç†æ•°æ®
#å½’çº¦æ­¥éª¤
def reduceAndCluster(df, clusters=3):
    #å±žæ€§è§„çº¦
    survivedSeries = pd.Series(df['Survived'],name='Survived')#å°†å€¼æ‹¿å‡ºæ¥
    X = df.values[:, 2::]
    y = df.values[:, 0]

    variance_pct = .99

    # Create PCA object
    pca = PCA(n_components=variance_pct)

    # Transform the initial features
    X_transformed = pca.fit_transform(X,y)

    # Create a data frame from the PCA'd data
    pcaDataFrame = pd.DataFrame(X_transformed)

    print("åŽŸæ•°æ®ç»´åº¦",X.shape[1])
    print("PCAåŽç»´åº¦",X_transformed.shape[1])


    #å€¼è§„çº¦ã€‚èšç±»
    from sklearn.cluster import KMeans

    #èšç±»å±žæ€§ æŽ¢æµ‹æ ·æœ¬é—´çš„ç›¸å…³æ€§
    kmeans = KMeans(n_clusters=clusters, random_state=np.random.RandomState(4), init='random')

    #æˆ‘ä»¬åˆ†ä¸ºè®­ç»ƒé›†å’Œæµ‹è¯•é›†
    split_train, split_cv = train_test_split(df, test_size=0.2, random_state=0)

    trainClusterIds = kmeans.fit_predict(X_transformed[:split_train.shape[0]])#å¾—åˆ°æ¯ä¸ªæ ·æœ¬çš„èšç±»ä¸­å¿ƒ
    print("è®­ç»ƒæ ·æœ¬çš„èšç±»ä¸­å¿ƒæ˜¯:",trainClusterIds)

    testClusterIds = kmeans.predict(X_transformed[split_train.shape[0]:])
    # print
    # "clusterIds shape for test data: ", testClusterIds.shape
    print ("æµ‹è¯•æ ·æœ¬çš„èšç±»ä¸­å¿ƒæ˜¯: ", testClusterIds)

    clusterIds = np.concatenate([trainClusterIds,testClusterIds])
    print("æ•´ä½“çš„æ ·æœ¬ä¸­å¿ƒ: ", clusterIds.shape)

    # åˆ›å»ºèšç±»ä¸­å¿ƒçš„Id
    clusterIdSeries = pd.Series(clusterIds, name='ClusterId')
    df = pd.concat([survivedSeries, clusterIdSeries, pcaDataFrame], axis=1)

    return df

#æ•°æ®è§„çº¦å’Œé™ä½Žç»´åº¦
def FeatureReduceCluster(df):
    #åˆ é™¤é«˜ç›¸å…³æ€§ç»´åº¦
    df = processDrops(df)
    #æ•°æ®è§„çº¦ç»´åº¦ PCA+ KMeans
    #ä½†ä»æŽ¥è¿‘äºŽä¿æŒåŽŸå§‹æ•°æ®çš„å®Œæ•´æ€§ã€‚ä¹Ÿå°±æ˜¯è¯´ï¼Œåœ¨å½’çº¦åŽçš„æ•°æ®é›†ä¸Šè¿›è¡Œæ•°æ®æŒ–æŽ˜å°†æ›´åŠ æœ‰æ•ˆï¼Œä»ç„¶äº§ç”Ÿå‡ ä¹Žç›¸åŒçš„æ•°æ®åˆ†æžç»“æžœã€‚
    #PCAé€šè¿‡åˆ›å»ºä¸€ä¸ªæ›¿æ¢çš„è¾ƒå°çš„å˜é‡é›†ç»„åˆå±žæ€§çš„åŸºæœ¬è¦ç´ 
    #è§„çº¦æœ‰å±žæ€§å’Œæ•°å€¼
    df = reduceAndCluster(df,clusters=3)

    return df
```

```python
#-*-coding:utf-8-*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import  preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
#ä¸€ä¸ªå±žæ€§ä¸€ä¸ªå±žæ€§çš„æŽ¢ç´¢ä»¥åŠå¤„ç†
#é€šè¿‡exploreæŽ¢æµ‹åˆ°å…·ä½“çš„æ•°å€¼å±žæ€§ä»¥åŠç¼ºçœå€¼å±žæ€§ä»¥åŠå­—ç¬¦å±žæ€§
#é¢˜æ„ä¸Šç»™å‡ºäº†such as women, children, and the upper-class ä¹Ÿå°±æ˜¯ç­‰çº§ï¼Œå¹´é¾„ï¼Œæ€§åˆ«éžå¸¸ç›¸å…³
#åŒæ—¶æŽ¢ç´¢åˆ°stdå˜æ¢è¾ƒå¤§çš„æ˜¯ageå’ŒFareå±žæ€§

#Pclass
#exploreåŽå¯èƒ½ç­‰çº§è¶Šé«˜èŽ·æ•‘çš„å¯èƒ½æ€§è¶Šå¤§
#ä½†æ˜¯ä¸èƒ½ç¡®å®šæ˜¯å¦ä¸€å®šæ‰€ä»¥æˆ‘ä»¬ç”¨è™šæ‹ŸåŒ–æ¥å¤„ç†one-hot_encoderç›¸å½“äºŽ

def processPclass(df):
    df.Pclass[df.Pclass.isnull()] = df.Pclass.dropna().mode().values
    dummies_df = pd.get_dummies(df['Pclass'],prefix='Pclass')#å‰ç¼€
    df = pd.concat([df,dummies_df],axis=1)
    return df

#Nameå±žæ€§
#æåˆ°çš„å…³é”®å±žæ€§æœ‰å¹´é¾„æ€§åˆ«ï¼Œç­‰çº§ å‡æœ‰è¯´æ˜Žæ•´ä¸ªå±žæ€§æ¯”è¾ƒé‡è¦
#é•¿åº¦é•¿è¯´æ˜Žè¿™ä¸ªäººå¯èƒ½åœ°ä½é«˜
#å­—ç¬¦ç±»åž‹å¯èƒ½éœ€è¦ç¦»æ•£åŒ–æˆ–è€…è™šæ‹ŸåŒ–
#ç”±äºŽå±žæ€§éœ€è¦å–å€¼ç§ç±»å¤ªå¤šã€‚æˆ‘ä»¬å°†æ‰€æœ‰æ›¿æ¢ä¸ºå‡ ä¸ªå…³é”®çš„å±žæ€§

def substrings_in_string(big_string, substrings):
    for substring in substrings:
        if big_string.find(substring)!=-1:
            return substring
    return np.nan

#æ›¿æ¢è§„åˆ™
def replace_titles(x):
    title = x['Title']
    if title in ['Mr', 'Don', 'Major', 'Capt', 'Jonkheer', 'Rev', 'Col']:
        return 'Mr'
    elif title in ['Master']:
        return 'Master'
    elif title in ['Countess', 'Mme', 'Mrs']:
        return 'Mrs'
    elif title in ['Mlle', 'Ms', 'Miss']:
        return 'Miss'
    elif title == 'Dr':
        if x['Sex'] == 'male':
            return 'Mr'
        else:
            return 'Mrs'
    elif title == '':
        if x['Sex'] == 'male':
            return 'Master'
        else:
            return 'Miss'
    else:
        return title
from  sklearn  import preprocessing
le = preprocessing.LabelEncoder()
def processName(df):
    # title_list = ['Mrs', 'Mr', 'Master', 'Miss', 'Major', 'Rev',
    #               'Dr', 'Ms', 'Mlle', 'Col', 'Capt', 'Mme', 'Countess',
    #               'Don', 'Jonkheer']
    # df['Title'] = df['Name'].map(lambda x:substrings_in_string(str(x),title_list))
    # #ä»ç„¶å¤ªå¤šç±»åž‹ï¼Œæˆ‘ä»¬ç®€åŒ–å°±4ç§ Mr Mrs Miss Master
    # df['Title'] = df.apply(replace_titles,axis=1)
    # #å­—ç¬¦ç±»åž‹ç¦»æ•£åŒ–
    # le.fit(df['Title'])
    # title_le = le.transform(df['Title'])
    # df['Title_id'] = title_le.astype(np.int32)
    df['Names'] = df['Name'].map(lambda x: len(re.split(' ', x)))  # è¿™ä¸ªåå­—æœ‰å‡ ä¸ªè¯ä¹Ÿå°±æ˜¯é•¿åº¦
    df['Title'] = df['Name'].map(lambda x: re.compile(",(.*?)\.").findall(x)[0])#å…¶ä¸­ç¬¬ä¸€ä¸ª.æ˜¯ä»»æ„å­—ç¬¦
    df['Title'] = df.apply(replace_titles,axis=1)
    # #å­—ç¬¦ç±»åž‹ç¦»æ•£åŒ–
    le.fit(df['Title'])
    title_le = le.transform(df['Title'])
    df['Title_id'] = title_le.astype(np.int32)
    return df


#Sexå±žæ€§
#æŽ¢ç´¢åŽå‘çŽ°å¥³æ€§å­˜æ´»çŽ‡é«˜å¤§æ¦‚ç™¾åˆ†ä¹‹75å·¦å³æ‰€ä»¥æ˜¯å…³é”®å±žæ€§
#æ–‡æœ¬-ã€‹æ•°å€¼ ç”±äºŽç”·å¥³æœ‰åŒºåˆ«æ‰€æœ‰è½¬æ¢æˆæ•°å€¼å°±è¡Œäº†
def processSex(df):
    df['Gender'] = df['Sex'].map({'female':1,'male':2}).astype(np.int32)
    return df


#Ageå±žæ€§
#stdå˜æ¢è¾ƒå¤§ã€‚ç¼ºçœå€¼æœ‰
#20-30æœ€å¤šäººç„¶åŽå°å­©ç¬¬äºŒå¤šã€‚æœ€å°‘çš„æ˜¯è€äºº
#å°å­©èŽ·æ•‘ä»¥åŠè€äººèŽ·æ•‘å‡ çŽ‡è¾ƒé«˜ã€‚å…¶ä»–å·®ä¸å¤šä¸€èˆ¬å·¦å³

def setMissingData(df,features=[],missFeature='Age'):
    feature_df = df[features]
    X = feature_df[df[missFeature].notnull()].as_matrix()[:,1::]
    y = feature_df[df[missFeature].notnull()].as_matrix()[:,0]
    rtr = RandomForestRegressor(n_estimators=2000,n_jobs=-1)#æ— é™åˆ¶å¤„ç†æœº
    rtr.fit(X,y)
    predicitedAges = rtr.predict(feature_df[df[missFeature].isnull()].as_matrix()[:,1:])
    df.loc[(df[missFeature].isnull()),missFeature] = predicitedAges
    return df

# def setMissingAges(df):
#
#     age_df = df[['Age', 'Embarked', 'Fare', 'Parch', 'SibSp', 'Title_id', 'Pclass', 'Names', 'CabinLetter']]
#     X = age_df.loc[(df.Age.notnull())].values[:, 1::]
#     y = age_df.loc[(df.Age.notnull())].values[:, 0]
#
#     rtr = RandomForestRegressor(n_estimators=2000, n_jobs=-1)
#     rtr.fit(X, y)
#
#     predictedAges = rtr.predict(age_df.loc[(df.Age.isnull())].values[:, 1::])
#     df.loc[(df.Age.isnull()), 'Age'] = predictedAges
#     return  df
def processAge(df):
    #å…ˆå¡«ç¼ºçœå€¼
    #é¢„æµ‹çš„æ–¹æ³•RandomForest
    df = setMissingData(df, features=['Age','Embarked','Fare', 'Parch', 'SibSp', 'Title_id','Pclass','Names','CabinLetter'], missFeature='Age')
    #df = setMissingAges(df)
    #æ­¤å¤„ç”¨ä¸­ä½æ•°ä»¥åŠå‡å€¼å¡«å……ä½†æ˜¯éœ€è¦å…ˆåˆ†å±‚å†æ±‚å‡å€¼ã€‚
    # mean_master = np.average(df['Age'][df.Title=='Master'].dropna())
    # mean_mr = np.average(df['Age'][df.Title=='Mr'].dropna())
    # mean_miss = np.average(df['Age'][df.Title=='Miss'].dropna())
    # mean_mrs = np.average(df['Age'][df.Title=='Mrs'].dropna())
    # df.loc[(df.Age.isnull())&(df.Title=='Master'),'Age'] = mean_master
    # df.loc[(df.Age.isnull()) & (df.Title == 'Mr'), 'Age'] = mean_mr
    # df.loc[(df.Age.isnull())&(df.Title=='Miss'),'Age'] = mean_miss
    # df.loc[(df.Age.isnull()) & (df.Title == 'Mrs'), 'Age'] = mean_mrs
    scaler = preprocessing.StandardScaler()
    df['Age_scaled'] = scaler.fit_transform(df['Age'])
    #ç‰¹åˆ«æåˆ°è€äººå°å­©ã€‚é‚£ä¹ˆæ˜¾ç„¶è¦ç¦»æ•£åŒ–å¹´é¾„
    # bin into quartiles and create binary features
    #æŒ‰ç…§é¢‘çŽ‡æŽ¥è¿‘çš„ç±»åˆ«ç¼–å·åœ¨ä¸€èµ·
    df['Age_bin'] = pd.qcut(df['Age'],4)
    #æ­¤æ—¶stdæ¯”è¾ƒå°äº†å¯ä»¥ä¸å˜æ¢
    df['Age_bin_id'] = pd.factorize(df['Age_bin'])[0]+1

    df['Child'] = (df['Age']<13).astype(int)

    #å˜åŒ–ä¸å¤§
    # from sklearn import  preprocessing
    # scaler = preprocessing.StandardScaler()
    # df['Age_bin_id_scaled'] = scaler.fit_transform(df['Age_bin_id'])
    return  df


#å¤„ç†å…„å¼Ÿå§å¦¹å’Œé…å¶çš„æ•°é‡
#è”æƒ³åˆ°å¦‚æžœè¿™ä¸ªæ•°ç›®è¾ƒå¤šçš„è¯å¯èƒ½æ˜¯å…„å¼Ÿçœ‹å¹´é¾„è¾ƒå°çš„è¯å°±æ˜¯å­©å­
#è€Œè¿™ä¸¤ä¸ªå±žæ€§æˆ‘ä»¬è”æƒ³åˆ°å…¶å®žå¯ä»¥èžåˆæˆä¸€ä¸ªå®¶åº­å±žæ€§ã€‚è€Œä¸”äººæ•°è¶Šå¤šæƒ…å†µæœ‰å¯èƒ½æ²¡èŽ·æ•‘
#0æ¯”è¾ƒå¤šæˆ‘ä»¬ç®—ä¸Šè‡ªå·±
def processSibsp(df):
    df['SibSp'] = df['SibSp'] + 1
    return df

def processParch(df):
    df['Parch'] = df['Parch'] + 1
    return df

def processFamily(df):
    df = processSibsp(df)
    df = processParch(df)
    df['Family'] = df['SibSp'] + df['Parch']
    return df

#å¤„ç†Fare stdå˜æ¢è¾ƒå¤§Scaledå¿…é¡»çš„ã€‚åŒæ—¶0æ¯”è¾ƒå¤šå¹³æ»‘åŒ–.è€Œå’Œå…³é”®å±žæ€§çŒœæƒ³å’Œupper classæœ‰å…³
#æ´»ä¸‹æ¥çš„äººï¼ŒFareæ€»èŠ±è´¹è¾ƒå¤š
#ç®—ä¸€ä¸‹æ¯ä¸ªäººçš„èŠ±è´¹
# çš„å¯èƒ½æ€§æ¯”è¾ƒå¤§å¯èƒ½äººç‰©è¶Šæœ‰
def processFare(df):
    #stdå¤„ç†æ–¹å¼æœ‰å¾ˆå¤šã€‚æœ€åŸºæœ¬æ˜¯scaled.è¿˜å¯ä»¥ç¦»æ•£åŒ–
    #å…ˆå¹³æ»‘åŒ–ä½¿å¾—0æ²¡æœ‰é‚£ä¹ˆå¤š
    df['Fare'][df.Fare.isnull()] = 0 #
    df.loc[(df.Fare==0),'Fare'] = df['Fare'][df.Fare.nonzero()[0]].min()/10
    sclar = preprocessing.StandardScaler()
    df['Fare_scaled'] = sclar.fit_transform(df['Fare'])
    df['Fare_bin'] = pd.qcut(df['Fare'],4)
    df['Fare_bin_id'] = pd.factorize(df['Fare_bin'])[0] + 1
    return df

#æˆ‘ä»¬çŒœæµ‹å’Œç­‰çº§æœ‰å…³ã€‚ç±»ä¼¼åå­—å¤„ç†æ–¹å¼æ— æ³•ç¦»æ•£åŒ–ã€‚åªæœ‰ç”¨å‡ ä¸ªä»£è¡¨çš„ç±»åž‹æ¥è¡¨ç¤º
#æˆ‘ä»¬éœ€è¦åˆ†æžç¥¨çš„å‰ç¼€åŒæ—¶å‘çŽ°å’ŒPclassçš„å…³ç³»ä¹Ÿå°±æ˜¯è¯´å‰ç¼€ç¡®å®žæœ‰ç­‰çº§çš„å…³ç³»
#è¿™ç§å­—ç¬¦ä¸²æˆ‘ä»¬éƒ½éœ€è¦å‰–æžå‰ç¼€å’ŒåŽç¼€ç„¶åŽç¼–ç 
#èŽ·å–å‰ç¼€
import re
def getTicketPrefix(ticket):
    match = re.compile("([a-zA-Z\.\/]+)").search(ticket)
    if match:#æœ‰å‰ç¼€
        return match.group()
    else:#æ²¡æœ‰å‰ç¼€
        return 'U'

#èŽ·å–ç¥¨çš„æ•°å­—
def getTicketNumber(ticket):
    match = re.compile("([0-9]+$)").search(ticket)
    if match:
        return  match.group()
    else:
        return '0'

def processTicket(df):
    #å…ˆåŒ–ä½œå­—ç¬¦ä¸²æ•°ç»„
    #strTick = df.Ticket.map(lambda x: str(x))
    #lenTick = df.Ticket.map(lambda x: len(x))
    df['TicketPrefix'] = df['Ticket'].map(lambda x: getTicketPrefix(x.upper()))
    df['TicketPrefix'] = df['TicketPrefix'].map(lambda x: re.sub('[\.?\/?]', '', x))  # åŽ»æŽ‰./
    df['TicketPrefix'] = df['TicketPrefix'].map(lambda x: re.sub('STON', 'SOTON', x))
    df['TicketPrefix'] = pd.factorize(df['TicketPrefix'])[0]
    df['TicketNumber'] = df['Ticket'].map(lambda x: getTicketNumber(x))
    df['TicketNumberLength'] = df['TicketNumber'].map(lambda x: len(x)). astype(int)
    df['TicketNumberStart'] = df['TicketNumber'].map(lambda x: x[0:1]).astype(int)
    df['TicketNumber'] = df['TicketNumber'].astype(int)
    return df


#Cabinæœ‰å¾ˆå¤šç¼ºçœå€¼
#å‘çŽ°æ˜¯å¦ç¼ºçœå¯¹ç»“æžœå½±å“å¾ˆå¤§
#åŒæ—¶å‚æ•°è´¡çŒ®åº¦ä¸å¤Ÿçš„è¯´æ˜Žåˆ’åˆ†ä¸å¤Ÿç»†
#èƒ½ä»£è¡¨ç€ä¸€äº›éšå«ä¿¡æ¯ã€‚æ¯”å¦‚èˆ¹èˆ±å·Cabinè¿™ä¸€å±žæ€§ï¼Œç¼ºå¤±å¯èƒ½ä»£è¡¨å¹¶æ²¡æœ‰èˆ¹èˆ±
#å‘çŽ°å­—æ¯è·Ÿç­‰çº§æœ‰å…³
def getCabinLetter(cabin):
    match = re.compile("([a-zA-Z]+)").search(cabin)
    if match:
        return  match.group()
    else:
        return 'U'

def getCabinNumber(cabin):
    match = re.compile("([0-9]+)").search(cabin)
    if match:
        return match.group()
    else:
        return 0

def processCabin(df):
    df['Cabin'][df.Cabin.isnull()] = 'U0'
    df['CabinLetter'] = df['Cabin'].map(lambda x:getCabinLetter(x))
    df['CabinLetter'] = pd.factorize(df['CabinLetter'])[0]
    df['CabinNumber'] = df['Cabin'].map(lambda x:getCabinNumber(x)).astype(int) + 1 #0å¤ªå¤š
    #stdæ¯”è¾ƒå¤§æ‰€ä»¥æˆ‘ä»¬è¦æ ‡å‡†åŒ–
    sclar = preprocessing.StandardScaler()
    df['CabinNumber'] = sclar.fit_transform(df['CabinNumber'])
    return df



#å¤„ç†Embarked ä¸Šèˆ¹ä½ç½®
#çŒœæƒ³ä¸Šèˆ¹çš„ä½ç½®é è¿‘æ•‘ç”Ÿå¤„ä»¥åŠæ˜¯ä»¥æœ€è¿‘ç­‰çº§é«˜çš„åœ°æ–¹æ‰€ä»¥è¿™ä¸ªå±žæ€§éœ€è¦
#Sçª—å£æœ€å¤šã€‚CèŽ·æ•‘å‡ çŽ‡æœ€å¤§
def processEmbarked(df):
    df.Embarked[df.Embarked.isnull()] = df.Embarked.dropna().mode()#çºµæ•°
    # #å¦‚æžœæ ¹æ®æ•°å­—å˜åŒ–æœ‰ä¸€å®šè¶‹åŠ¿ã€‚
    # df.Embarked = pd.factorize(df['Embarked'])[0]#æ•°å€¼åŒ–
    # #ä¹Ÿå¯ä»¥è™šæ‹ŸåŒ–
    df = pd.concat([df,pd.get_dummies(df['Embarked'],prefix='Embarked')],axis=1)
    df.Embarked = pd.factorize(df['Embarked'])[0]
    return df



#è€ƒè™‘ç»„åˆç‰¹å¾
#è€Œæˆ‘ä»¬ç»„åˆç‰¹å¾å…¶å®žæ˜¯æŠŠæ¯ä¸€ç»´ç‰¹å¾æƒé‡åº”è¯¥ç›¸ç­‰çœ‹å¾…ã€‚æ‰€ä»¥æœ€å¥½éƒ½scaled
#ç„¶åŽæˆ‘ä»¬éåŽ†æ‰€æœ‰å¯èƒ½çš„ç»„åˆ
def combineFeature(df):
    print("Starting With",df.columns.size,"æ‰‹åŠ¨ç”Ÿæˆç»„åˆç‰¹å¾",df.columns.values)
    #åªè€ƒè™‘è¿žç»­å±žæ€§
    numerics = df.loc[:,['Age_scaled','Fare_scaled','Pclass','Parch','SibSp',
                         'Names','CabinNumber','Age_bin_id','Fare_bin_id']]
    print("\nFeatures used for automated feature generation:\n",numerics.head(10))

    new_fields_count = 0
    for i in range(0,numerics.columns.size - 1):
        for j in range(0,numerics.columns.size-1):
            if i<=j:
                name = str(numerics.columns.values[i]) + "*" + str(numerics.columns.values[j])
                df = pd.concat([df,pd.Series(numerics.iloc[:,i]*numerics.iloc[:,j],name=name)],axis=1)
                new_fields_count+=1
            if i<j:
                name = str(numerics.columns.values[i]) + "+" + str(numerics.columns.values[j])
                df = pd.concat([df, pd.Series(numerics.iloc[:, i] + numerics.iloc[:, j], name=name)], axis=1)
                new_fields_count += 1
            if not i == j:
                name = str(numerics.columns.values[i]) + "/" + str(numerics.columns.values[j])
                df = pd.concat([df, pd.Series(numerics.iloc[:, i] / numerics.iloc[:, j], name=name)], axis=1)
                name = str(numerics.columns.values[i]) + "-" + str(numerics.columns.values[j])
                df = pd.concat([df, pd.Series(numerics.iloc[:, i] - numerics.iloc[:, j], name=name)], axis=1)
                new_fields_count += 2
    print("\n",new_fields_count,"new features generated")
    return df

#æ•°æ®è§„çº¦å’Œé™ä½Žç»´åº¦
# æ•°æ®è§„çº¦ç»´åº¦ PCA+ KMeans
# ä½†ä»æŽ¥è¿‘äºŽä¿æŒåŽŸå§‹æ•°æ®çš„å®Œæ•´æ€§ã€‚ä¹Ÿå°±æ˜¯è¯´ï¼Œåœ¨å½’çº¦åŽçš„æ•°æ®é›†ä¸Šè¿›è¡Œæ•°æ®æŒ–æŽ˜å°†æ›´åŠ æœ‰æ•ˆï¼Œä»ç„¶äº§ç”Ÿå‡ ä¹Žç›¸åŒçš„æ•°æ®åˆ†æžç»“æžœã€‚
# PCAé€šè¿‡åˆ›å»ºä¸€ä¸ªæ›¿æ¢çš„è¾ƒå°çš„å˜é‡é›†ç»„åˆå±žæ€§çš„åŸºæœ¬è¦ç´ 
# è§„çº¦æœ‰å±žæ€§å’Œæ•°å€¼

#ç‰¹å¾æå–é™ä½Žç»´åº¦
def processDrops(df):
    # global df
    # rawDropList = ['Name', 'Names', 'Title', 'Sex', 'SibSp', 'Parch', 'Pclass', 'Embarked', \
    #                'Cabin', 'CabinLetter', 'CabinNumber', 'Age', 'Fare', 'Ticket', 'TicketNumber']
    #stringsDropList = ['Title', 'Name', 'Cabin', 'Ticket', 'Sex', 'Ticket', 'TicketNumber']
    # # if not keep_raw:
    # #     df.drop(rawDropList, axis=1, inplace=True)
    # # # elif not keep_strings:
    # #     df.drop(stringsDropList, axis=1, inplace=True)
    # df.drop(stringsDropList,axis=1,inplace=True)
    #ls = df.loc[:,[['PassengerId','Survived']]]
    # Ps = df.loc[:, ['PassengerId', 'Survived']]
    # Ps
    # Ps.corr()
    #åˆ é™¤å­—ç¬¦å±žæ€§çš„å€¼
    stringsDropList = ['Title', 'Name', 'Cabin', 'Ticket', 'Sex', 'Ticket', 'TicketNumber','Age_bin','Fare_bin']
    df.drop(stringsDropList,axis=1,inplace=True)

    #å‘çŽ°æŽ’é™¤survived å’ŒpassengerIdä»¥åŽ
    df_corr = df.drop(['Survived', 'PassengerId'], axis=1).corr(method='spearman')
    mask = np.ones(df_corr.columns.size) - np.eye(df_corr.columns.size)#é»˜è®¤ä¸­é—´ä¸º0ç›¸å½“äºŽè·Ÿè‡ªå·±æ— å…³
    df_corr = mask * df_corr#ä¹˜æ˜¯å¯¹åº”å…ƒç´ ä¹˜
    #å¾—åˆ°æ•´ä½“çš„åæ–¹å·®çŸ©é˜µ

    #èµ›é€‰å‡ºç›¸å…³æ€§å¤ªé«˜çš„å±žæ€§å†—æ‚å±žæ€§
    drops = []
    for col in df_corr.columns.values:
        if np.in1d([col],drops):#å·²ç»æœ‰äº†
            continue
        corr = df_corr[abs(df_corr[col]) > 0.98].index#ç¬¬colåˆ—ä¸­ç›¸å…³æ€§å¤ªå¤§çš„èˆå¼ƒ
        drops = np.union1d(drops, corr)
    print("\nDropping", drops.shape[0], "highly correlated features...\n")# , drops
    df.drop(drops,axis=1,inplace=True)
    return  df

#æˆ‘ä»¬å¾—åˆ°äº†å…·æœ‰å¤§é‡ç‰¹å¾çš„ç»´åº¦å¾ˆé«˜çš„æ•°æ®é›†ï¼Œç‰¹å¾è¾ƒå¤šä¸èƒ½ç›´æŽ¥ç”¨æ¥ä½œä¸ºæ¨¡åž‹è¾“å…¥ï¼Œä¸€æ˜¯å› ä¸ºè¿™äº›ç‰¹å¾é—´å…·æœ‰å¤šé‡å…±çº¿æ€§ï¼Œ
# å¯èƒ½ ä¼šå¯¼è‡´ç©ºé—´çš„ä¸ç¨³å®šï¼›äºŒæ˜¯å› ä¸ºé«˜ç»´ç©ºé—´æœ¬èº«å…·æœ‰ç¨€ç–æ€§ï¼Œä¸€ç»´æ­£æ€åˆ†å¸ƒæœ‰68%çš„å€¼è½äºŽæ­£è´Ÿæ ‡å‡†å·®ä¹‹é—´ï¼Œè€Œåœ¨åç»´ç©ºé—´ä¸Šåªæœ‰0.02%ï¼›ä¸‰æ˜¯ç”±äºŽè¿‡å¤šçš„å±žæ€§
# ä¼šä½¿æŒ–æŽ˜éœ€è¦å¾ˆé•¿æ—¶é—´ã€‚å¯¹äºŽä¸€äº›æ¨¡åž‹æ¥è¯´ï¼Œæ¯”å¦‚ä½¿ç”¨L1ï¼ˆLassoï¼‰ï¼Œå½“æœ‰å¤§é‡å±žæ€§æ—¶æ•ˆæžœå¾ˆå¥½ï¼Œå› ä¸ºå®ƒå¯ä»¥æœ‰æ•ˆå¿½ç•¥æŽ‰å™ªå£°å˜é‡ã€‚è€Œä¸€äº›æ¨¡åž‹åˆ™å®¹æ˜“è¿‡æ‹Ÿ åˆã€‚
#ä¼šä½¿ç”¨èšç±»åˆ†æžå’Œé™ç»´æ¥å¤„ç†æ•°æ®
#å½’çº¦æ­¥éª¤
def reduceAndCluster(df,input_df,clusters=3):
    #å±žæ€§è§„çº¦
    survivedSeries = pd.Series(df['Survived'],name='Survived')#å°†å€¼æ‹¿å‡ºæ¥
    df.drop('Survived',axis=1,inplace=True)
    df = df.reindex_axis(input_df.columns,axis=1)#é‡æ–°æŒ‰ç…§input_dfåˆ—æ–¹å¼
    X = df.values[:, 1::]
    y = df.values[:, 0]#ç±»åˆ«å±žæ€§

    variance_pct = .99

    # Create PCA object
    pca = PCA(n_components=variance_pct)

    # Transform the initial features
    X_transformed = pca.fit_transform(X,y)

    # Create a data frame from the PCA'd data
    pcaDataFrame = pd.DataFrame(X_transformed)

    print("åŽŸæ•°æ®ç»´åº¦",X.shape[1])
    print("PCAåŽç»´åº¦",X_transformed.shape[1])


    #å€¼è§„çº¦ã€‚èšç±»
    from sklearn.cluster import KMeans

    #èšç±»å±žæ€§ æŽ¢æµ‹æ ·æœ¬é—´çš„ç›¸å…³æ€§
    kmeans = KMeans(n_clusters=clusters, random_state=np.random.RandomState(4), init='random')

    #æˆ‘ä»¬åˆ†ä¸ºè®­ç»ƒé›†å’Œæµ‹è¯•é›†
    #split_train, split_cv = train_test_split(df, test_size=0.2, random_state=0)

    trainClusterIds = kmeans.fit_predict(X_transformed[:input_df.shape[0]])#å¾—åˆ°æ¯ä¸ªæ ·æœ¬çš„èšç±»ä¸­å¿ƒ
    print("è®­ç»ƒæ ·æœ¬çš„èšç±»ä¸­å¿ƒæ˜¯:",trainClusterIds)

    testClusterIds = kmeans.predict(X_transformed[input_df.shape[0]:])
    # print
    # "clusterIds shape for test data: ", testClusterIds.shape
    print ("æµ‹è¯•æ ·æœ¬çš„èšç±»ä¸­å¿ƒæ˜¯: ", testClusterIds)

    clusterIds = np.concatenate([trainClusterIds,testClusterIds])
    print("æ•´ä½“çš„æ ·æœ¬ä¸­å¿ƒ: ", clusterIds.shape)

    # åˆ›å»ºèšç±»ä¸­å¿ƒçš„Id
    clusterIdSeries = pd.Series(clusterIds, name='ClusterId')
    df = pd.concat([survivedSeries, clusterIdSeries, pcaDataFrame], axis=1)

    return df



#æ•°æ®æ¸…ç†ä»¥åŠç‰¹å¾å·¥ç¨‹
def clean_and_feature_engineer(df,input_df):
    df = processCabin(df)
    df = processEmbarked(df)
    df = processTicket(df)
    df = processFare(df)
    df = processFamily(df)
    df = processName(df)
    df = processSex(df)
    df = processPclass(df)
    df = processAge(df)
    #ç»„åˆç‰¹å¾
    df = combineFeature(df)

    #åˆ é™¤å†—æ‚å±žæ€§
    df = processDrops(df)
    input_df = df[:input_df.shape[0]]
    submit_df = df[input_df.shape[0]:]
    drop_list = ['PassengerId']
    input_df.drop(drop_list,axis=1,inplace=1)
    submit_df.drop(drop_list,axis=1,inplace=1)

    #è§„çº¦
    df = reduceAndCluster(df,input_df)

    return df


# #ä¸»å‡½æ•°
def dataprocess(filename='train.csv',testname='test.csv'):
#     #å°†è®­ç»ƒå’Œæµ‹è¯•ç»„åˆ
#é‡æ–°ç´¢å¼•ä»¥åŠåŽ»æŽ‰NANçš„survivedæäº¤å˜é‡
    # å°†è®­ç»ƒå’Œæµ‹è¯•ç»„åˆ
    input_df = pd.read_csv(filename)
    submit_df = pd.read_csv(testname)
    df = pd.concat([input_df, submit_df])
    df.reset_index(inplace=True)
    df.drop('index', axis=1, inplace=True)

    df = clean_and_feature_engineer(df,input_df,submit_df)
    input_df = df[:input_df.shape[0]]
    submit_df = df[input_df.shape[0]:]
    submit_df.reset_index(inplace=True)
    submit_df.drop('index', axis=1, inplace=True)
    submit_df.drop('Survived', axis=1, inplace=1)
    return input_df,submit_df
#     return data_train
```

```python
#-*-coding:utf-8-*-
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import  preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
#ä¸€ä¸ªå±žæ€§ä¸€ä¸ªå±žæ€§çš„æŽ¢ç´¢ä»¥åŠå¤„ç†
#é€šè¿‡exploreæŽ¢æµ‹åˆ°å…·ä½“çš„æ•°å€¼å±žæ€§ä»¥åŠç¼ºçœå€¼å±žæ€§ä»¥åŠå­—ç¬¦å±žæ€§
#é¢˜æ„ä¸Šç»™å‡ºäº†such as women, children, and the upper-class ä¹Ÿå°±æ˜¯ç­‰çº§ï¼Œå¹´é¾„ï¼Œæ€§åˆ«éžå¸¸ç›¸å…³
#åŒæ—¶æŽ¢ç´¢åˆ°stdå˜æ¢è¾ƒå¤§çš„æ˜¯ageå’ŒFareå±žæ€§

#Pclass
#exploreåŽå¯èƒ½ç­‰çº§è¶Šé«˜èŽ·æ•‘çš„å¯èƒ½æ€§è¶Šå¤§
#ä½†æ˜¯ä¸èƒ½ç¡®å®šæ˜¯å¦ä¸€å®šæ‰€ä»¥æˆ‘ä»¬ç”¨è™šæ‹ŸåŒ–æ¥å¤„ç†one-hot_encoderç›¸å½“äºŽ

def processPclass(df):
    df.Pclass[df.Pclass.isnull()] = df.Pclass.dropna().mode().values
    dummies_df = pd.get_dummies(df['Pclass'],prefix='Pclass')#å‰ç¼€
    df = pd.concat([df,dummies_df],axis=1)
    return df

#Nameå±žæ€§
#æåˆ°çš„å…³é”®å±žæ€§æœ‰å¹´é¾„æ€§åˆ«ï¼Œç­‰çº§ å‡æœ‰è¯´æ˜Žæ•´ä¸ªå±žæ€§æ¯”è¾ƒé‡è¦
#é•¿åº¦é•¿è¯´æ˜Žè¿™ä¸ªäººå¯èƒ½åœ°ä½é«˜
#å­—ç¬¦ç±»åž‹å¯èƒ½éœ€è¦ç¦»æ•£åŒ–æˆ–è€…è™šæ‹ŸåŒ–
#ç”±äºŽå±žæ€§éœ€è¦å–å€¼ç§ç±»å¤ªå¤šã€‚æˆ‘ä»¬å°†æ‰€æœ‰æ›¿æ¢ä¸ºå‡ ä¸ªå…³é”®çš„å±žæ€§

def substrings_in_string(big_string, substrings):
    for substring in substrings:
        if big_string.find(substring)!=-1:
            return substring
    return np.nan

#æ›¿æ¢è§„åˆ™
def replace_titles(x):
    title = x['Title']
    if title in ['Mr', 'Don', 'Major', 'Capt', 'Jonkheer', 'Rev', 'Col']:
        return 'Mr'
    elif title in ['Master']:
        return 'Master'
    elif title in ['Countess', 'Mme', 'Mrs']:
        return 'Mrs'
    elif title in ['Mlle', 'Ms', 'Miss']:
        return 'Miss'
    elif title == 'Dr':
        if x['Sex'] == 'male':
            return 'Mr'
        else:
            return 'Mrs'
    elif title == '':
        if x['Sex'] == 'male':
            return 'Master'
        else:
            return 'Miss'
    else:
        return title
from  sklearn  import preprocessing
le = preprocessing.LabelEncoder()
def processName(df):
    # title_list = ['Mrs', 'Mr', 'Master', 'Miss', 'Major', 'Rev',
    #               'Dr', 'Ms', 'Mlle', 'Col', 'Capt', 'Mme', 'Countess',
    #               'Don', 'Jonkheer']
    # df['Title'] = df['Name'].map(lambda x:substrings_in_string(str(x),title_list))
    # #ä»ç„¶å¤ªå¤šç±»åž‹ï¼Œæˆ‘ä»¬ç®€åŒ–å°±4ç§ Mr Mrs Miss Master
    # df['Title'] = df.apply(replace_titles,axis=1)
    # #å­—ç¬¦ç±»åž‹ç¦»æ•£åŒ–
    # le.fit(df['Title'])
    # title_le = le.transform(df['Title'])
    # df['Title_id'] = title_le.astype(np.int32)
    df['Names'] = df['Name'].map(lambda x: len(re.split(' ', x)))  # è¿™ä¸ªåå­—æœ‰å‡ ä¸ªè¯ä¹Ÿå°±æ˜¯é•¿åº¦
    df['Title'] = df['Name'].map(lambda x: re.compile(",(.*?)\.").findall(x)[0])#å…¶ä¸­ç¬¬ä¸€ä¸ª.æ˜¯ä»»æ„å­—ç¬¦
    df['Title'] = df.apply(replace_titles,axis=1)
    # #å­—ç¬¦ç±»åž‹ç¦»æ•£åŒ–
    le.fit(df['Title'])
    title_le = le.transform(df['Title'])
    df['Title_id'] = title_le.astype(np.int32)
    return df


#Sexå±žæ€§
#æŽ¢ç´¢åŽå‘çŽ°å¥³æ€§å­˜æ´»çŽ‡é«˜å¤§æ¦‚ç™¾åˆ†ä¹‹75å·¦å³æ‰€ä»¥æ˜¯å…³é”®å±žæ€§
#æ–‡æœ¬-ã€‹æ•°å€¼ ç”±äºŽç”·å¥³æœ‰åŒºåˆ«æ‰€æœ‰è½¬æ¢æˆæ•°å€¼å°±è¡Œäº†
def processSex(df):
    df['Gender'] = df['Sex'].map({'female':1,'male':2}).astype(np.int32)
    return df


#Ageå±žæ€§
#stdå˜æ¢è¾ƒå¤§ã€‚ç¼ºçœå€¼æœ‰
#20-30æœ€å¤šäººç„¶åŽå°å­©ç¬¬äºŒå¤šã€‚æœ€å°‘çš„æ˜¯è€äºº
#å°å­©èŽ·æ•‘ä»¥åŠè€äººèŽ·æ•‘å‡ çŽ‡è¾ƒé«˜ã€‚å…¶ä»–å·®ä¸å¤šä¸€èˆ¬å·¦å³

def setMissingData(df,features=[],missFeature='Age'):
    feature_df = df[features]
    X = feature_df[df[missFeature].notnull()].as_matrix()[:,1::]
    y = feature_df[df[missFeature].notnull()].as_matrix()[:,0]
    rtr = RandomForestRegressor(n_estimators=2000,n_jobs=-1)#æ— é™åˆ¶å¤„ç†æœº
    rtr.fit(X,y)
    predicitedAges = rtr.predict(feature_df[df[missFeature].isnull()].as_matrix()[:,1:])
    df.loc[(df[missFeature].isnull()),missFeature] = predicitedAges
    return df

# def setMissingAges(df):
#
#     age_df = df[['Age', 'Embarked', 'Fare', 'Parch', 'SibSp', 'Title_id', 'Pclass', 'Names', 'CabinLetter']]
#     X = age_df.loc[(df.Age.notnull())].values[:, 1::]
#     y = age_df.loc[(df.Age.notnull())].values[:, 0]
#
#     rtr = RandomForestRegressor(n_estimators=2000, n_jobs=-1)
#     rtr.fit(X, y)
#
#     predictedAges = rtr.predict(age_df.loc[(df.Age.isnull())].values[:, 1::])
#     df.loc[(df.Age.isnull()), 'Age'] = predictedAges
#     return  df
def processAge(df):
    #å…ˆå¡«ç¼ºçœå€¼
    #é¢„æµ‹çš„æ–¹æ³•RandomForest
    df = setMissingData(df, features=['Age','Embarked','Fare', 'Parch', 'SibSp', 'Title_id','Pclass','Names','CabinLetter'], missFeature='Age')
    #df = setMissingAges(df)
    #æ­¤å¤„ç”¨ä¸­ä½æ•°ä»¥åŠå‡å€¼å¡«å……ä½†æ˜¯éœ€è¦å…ˆåˆ†å±‚å†æ±‚å‡å€¼ã€‚
    # mean_master = np.average(df['Age'][df.Title=='Master'].dropna())
    # mean_mr = np.average(df['Age'][df.Title=='Mr'].dropna())
    # mean_miss = np.average(df['Age'][df.Title=='Miss'].dropna())
    # mean_mrs = np.average(df['Age'][df.Title=='Mrs'].dropna())
    # df.loc[(df.Age.isnull())&(df.Title=='Master'),'Age'] = mean_master
    # df.loc[(df.Age.isnull()) & (df.Title == 'Mr'), 'Age'] = mean_mr
    # df.loc[(df.Age.isnull())&(df.Title=='Miss'),'Age'] = mean_miss
    # df.loc[(df.Age.isnull()) & (df.Title == 'Mrs'), 'Age'] = mean_mrs
    scaler = preprocessing.StandardScaler()
    df['Age_scaled'] = scaler.fit_transform(df['Age'])
    #ç‰¹åˆ«æåˆ°è€äººå°å­©ã€‚é‚£ä¹ˆæ˜¾ç„¶è¦ç¦»æ•£åŒ–å¹´é¾„
    # bin into quartiles and create binary features
    #æŒ‰ç…§é¢‘çŽ‡æŽ¥è¿‘çš„ç±»åˆ«ç¼–å·åœ¨ä¸€èµ·
    df['Age_bin'] = pd.qcut(df['Age'],4)
    #æ­¤æ—¶stdæ¯”è¾ƒå°äº†å¯ä»¥ä¸å˜æ¢
    df['Age_bin_id'] = pd.factorize(df['Age_bin'])[0]+1

    df['Child'] = (df['Age']<13).astype(int)

    #å˜åŒ–ä¸å¤§
    # from sklearn import  preprocessing
    # scaler = preprocessing.StandardScaler()
    # df['Age_bin_id_scaled'] = scaler.fit_transform(df['Age_bin_id'])
    return  df


#å¤„ç†å…„å¼Ÿå§å¦¹å’Œé…å¶çš„æ•°é‡
#è”æƒ³åˆ°å¦‚æžœè¿™ä¸ªæ•°ç›®è¾ƒå¤šçš„è¯å¯èƒ½æ˜¯å…„å¼Ÿçœ‹å¹´é¾„è¾ƒå°çš„è¯å°±æ˜¯å­©å­
#è€Œè¿™ä¸¤ä¸ªå±žæ€§æˆ‘ä»¬è”æƒ³åˆ°å…¶å®žå¯ä»¥èžåˆæˆä¸€ä¸ªå®¶åº­å±žæ€§ã€‚è€Œä¸”äººæ•°è¶Šå¤šæƒ…å†µæœ‰å¯èƒ½æ²¡èŽ·æ•‘
#0æ¯”è¾ƒå¤šæˆ‘ä»¬ç®—ä¸Šè‡ªå·±
def processSibsp(df):
    df['SibSp'] = df['SibSp'] + 1
    return df

def processParch(df):
    df['Parch'] = df['Parch'] + 1
    return df

def processFamily(df):
    df = processSibsp(df)
    df = processParch(df)
    df['Family'] = df['SibSp'] + df['Parch']
    return df

#å¤„ç†Fare stdå˜æ¢è¾ƒå¤§Scaledå¿…é¡»çš„ã€‚åŒæ—¶0æ¯”è¾ƒå¤šå¹³æ»‘åŒ–.è€Œå’Œå…³é”®å±žæ€§çŒœæƒ³å’Œupper classæœ‰å…³
#æ´»ä¸‹æ¥çš„äººï¼ŒFareæ€»èŠ±è´¹è¾ƒå¤š
#ç®—ä¸€ä¸‹æ¯ä¸ªäººçš„èŠ±è´¹
# çš„å¯èƒ½æ€§æ¯”è¾ƒå¤§å¯èƒ½äººç‰©è¶Šæœ‰
def processFare(df):
    #stdå¤„ç†æ–¹å¼æœ‰å¾ˆå¤šã€‚æœ€åŸºæœ¬æ˜¯scaled.è¿˜å¯ä»¥ç¦»æ•£åŒ–
    #å…ˆå¹³æ»‘åŒ–ä½¿å¾—0æ²¡æœ‰é‚£ä¹ˆå¤š
    df['Fare'][df.Fare.isnull()] = 0 #
    df.loc[(df.Fare==0),'Fare'] = df['Fare'][df.Fare.nonzero()[0]].min()/10
    sclar = preprocessing.StandardScaler()
    df['Fare_scaled'] = sclar.fit_transform(df['Fare'])
    df['Fare_bin'] = pd.qcut(df['Fare'],4)
    df['Fare_bin_id'] = pd.factorize(df['Fare_bin'])[0] + 1
    return df

#æˆ‘ä»¬çŒœæµ‹å’Œç­‰çº§æœ‰å…³ã€‚ç±»ä¼¼åå­—å¤„ç†æ–¹å¼æ— æ³•ç¦»æ•£åŒ–ã€‚åªæœ‰ç”¨å‡ ä¸ªä»£è¡¨çš„ç±»åž‹æ¥è¡¨ç¤º
#æˆ‘ä»¬éœ€è¦åˆ†æžç¥¨çš„å‰ç¼€åŒæ—¶å‘çŽ°å’ŒPclassçš„å…³ç³»ä¹Ÿå°±æ˜¯è¯´å‰ç¼€ç¡®å®žæœ‰ç­‰çº§çš„å…³ç³»
#è¿™ç§å­—ç¬¦ä¸²æˆ‘ä»¬éƒ½éœ€è¦å‰–æžå‰ç¼€å’ŒåŽç¼€ç„¶åŽç¼–ç 
#èŽ·å–å‰ç¼€
import re
def getTicketPrefix(ticket):
    match = re.compile("([a-zA-Z\.\/]+)").search(ticket)
    if match:#æœ‰å‰ç¼€
        return match.group()
    else:#æ²¡æœ‰å‰ç¼€
        return 'U'

#èŽ·å–ç¥¨çš„æ•°å­—
def getTicketNumber(ticket):
    match = re.compile("([0-9]+$)").search(ticket)
    if match:
        return  match.group()
    else:
        return '0'

def processTicket(df):
    #å…ˆåŒ–ä½œå­—ç¬¦ä¸²æ•°ç»„
    #strTick = df.Ticket.map(lambda x: str(x))
    #lenTick = df.Ticket.map(lambda x: len(x))
    df['TicketPrefix'] = df['Ticket'].map(lambda x: getTicketPrefix(x.upper()))
    df['TicketPrefix'] = df['TicketPrefix'].map(lambda x: re.sub('[\.?\/?]', '', x))  # åŽ»æŽ‰./
    df['TicketPrefix'] = df['TicketPrefix'].map(lambda x: re.sub('STON', 'SOTON', x))
    df['TicketPrefix'] = pd.factorize(df['TicketPrefix'])[0]
    df['TicketNumber'] = df['Ticket'].map(lambda x: getTicketNumber(x))
    df['TicketNumberLength'] = df['TicketNumber'].map(lambda x: len(x)). astype(int)
    df['TicketNumberStart'] = df['TicketNumber'].map(lambda x: x[0:1]).astype(int)
    df['TicketNumber'] = df['TicketNumber'].astype(int)
    return df


#Cabinæœ‰å¾ˆå¤šç¼ºçœå€¼
#å‘çŽ°æ˜¯å¦ç¼ºçœå¯¹ç»“æžœå½±å“å¾ˆå¤§
#åŒæ—¶å‚æ•°è´¡çŒ®åº¦ä¸å¤Ÿçš„è¯´æ˜Žåˆ’åˆ†ä¸å¤Ÿç»†
#èƒ½ä»£è¡¨ç€ä¸€äº›éšå«ä¿¡æ¯ã€‚æ¯”å¦‚èˆ¹èˆ±å·Cabinè¿™ä¸€å±žæ€§ï¼Œç¼ºå¤±å¯èƒ½ä»£è¡¨å¹¶æ²¡æœ‰èˆ¹èˆ±
#å‘çŽ°å­—æ¯è·Ÿç­‰çº§æœ‰å…³
def getCabinLetter(cabin):
    match = re.compile("([a-zA-Z]+)").search(cabin)
    if match:
        return  match.group()
    else:
        return 'U'

def getCabinNumber(cabin):
    match = re.compile("([0-9]+)").search(cabin)
    if match:
        return match.group()
    else:
        return 0

def processCabin(df):
    df['Cabin'][df.Cabin.isnull()] = 'U0'
    df['CabinLetter'] = df['Cabin'].map(lambda x:getCabinLetter(x))
    df['CabinLetter'] = pd.factorize(df['CabinLetter'])[0]
    df['CabinNumber'] = df['Cabin'].map(lambda x:getCabinNumber(x)).astype(int) + 1 #0å¤ªå¤š
    #stdæ¯”è¾ƒå¤§æ‰€ä»¥æˆ‘ä»¬è¦æ ‡å‡†åŒ–
    sclar = preprocessing.StandardScaler()
    df['CabinNumber'] = sclar.fit_transform(df['CabinNumber'])
    return df



#å¤„ç†Embarked ä¸Šèˆ¹ä½ç½®
#çŒœæƒ³ä¸Šèˆ¹çš„ä½ç½®é è¿‘æ•‘ç”Ÿå¤„ä»¥åŠæ˜¯ä»¥æœ€è¿‘ç­‰çº§é«˜çš„åœ°æ–¹æ‰€ä»¥è¿™ä¸ªå±žæ€§éœ€è¦
#Sçª—å£æœ€å¤šã€‚CèŽ·æ•‘å‡ çŽ‡æœ€å¤§
def processEmbarked(df):
    df.Embarked[df.Embarked.isnull()] = df.Embarked.dropna().mode()#çºµæ•°
    # #å¦‚æžœæ ¹æ®æ•°å­—å˜åŒ–æœ‰ä¸€å®šè¶‹åŠ¿ã€‚
    # df.Embarked = pd.factorize(df['Embarked'])[0]#æ•°å€¼åŒ–
    # #ä¹Ÿå¯ä»¥è™šæ‹ŸåŒ–
    df = pd.concat([df,pd.get_dummies(df['Embarked'],prefix='Embarked')],axis=1)
    df.Embarked = pd.factorize(df['Embarked'])[0]
    return df



#è€ƒè™‘ç»„åˆç‰¹å¾
#è€Œæˆ‘ä»¬ç»„åˆç‰¹å¾å…¶å®žæ˜¯æŠŠæ¯ä¸€ç»´ç‰¹å¾æƒé‡åº”è¯¥ç›¸ç­‰çœ‹å¾…ã€‚æ‰€ä»¥æœ€å¥½éƒ½scaled
#ç„¶åŽæˆ‘ä»¬éåŽ†æ‰€æœ‰å¯èƒ½çš„ç»„åˆ
def combineFeature(df):
    print("Starting With",df.columns.size,"æ‰‹åŠ¨ç”Ÿæˆç»„åˆç‰¹å¾",df.columns.values)
    #åªè€ƒè™‘è¿žç»­å±žæ€§
    numerics = df.loc[:,['Age_scaled','Fare_scaled','Pclass','Parch','SibSp',
                         'Names','CabinNumber','Age_bin_id','Fare_bin_id']]
    print("\nFeatures used for automated feature generation:\n",numerics.head(10))

    new_fields_count = 0
    for i in range(0,numerics.columns.size - 1):
        for j in range(0,numerics.columns.size-1):
            if i<=j:
                name = str(numerics.columns.values[i]) + "*" + str(numerics.columns.values[j])
                df = pd.concat([df,pd.Series(numerics.iloc[:,i]*numerics.iloc[:,j],name=name)],axis=1)
                new_fields_count+=1
            if i<j:
                name = str(numerics.columns.values[i]) + "+" + str(numerics.columns.values[j])
                df = pd.concat([df, pd.Series(numerics.iloc[:, i] + numerics.iloc[:, j], name=name)], axis=1)
                new_fields_count += 1
            if not i == j:
                name = str(numerics.columns.values[i]) + "/" + str(numerics.columns.values[j])
                df = pd.concat([df, pd.Series(numerics.iloc[:, i] / numerics.iloc[:, j], name=name)], axis=1)
                name = str(numerics.columns.values[i]) + "-" + str(numerics.columns.values[j])
                df = pd.concat([df, pd.Series(numerics.iloc[:, i] - numerics.iloc[:, j], name=name)], axis=1)
                new_fields_count += 2
    print("\n",new_fields_count,"new features generated")
    return df

#æ•°æ®è§„çº¦å’Œé™ä½Žç»´åº¦
# æ•°æ®è§„çº¦ç»´åº¦ PCA+ KMeans
# ä½†ä»æŽ¥è¿‘äºŽä¿æŒåŽŸå§‹æ•°æ®çš„å®Œæ•´æ€§ã€‚ä¹Ÿå°±æ˜¯è¯´ï¼Œåœ¨å½’çº¦åŽçš„æ•°æ®é›†ä¸Šè¿›è¡Œæ•°æ®æŒ–æŽ˜å°†æ›´åŠ æœ‰æ•ˆï¼Œä»ç„¶äº§ç”Ÿå‡ ä¹Žç›¸åŒçš„æ•°æ®åˆ†æžç»“æžœã€‚
# PCAé€šè¿‡åˆ›å»ºä¸€ä¸ªæ›¿æ¢çš„è¾ƒå°çš„å˜é‡é›†ç»„åˆå±žæ€§çš„åŸºæœ¬è¦ç´ 
# è§„çº¦æœ‰å±žæ€§å’Œæ•°å€¼

#ç‰¹å¾æå–é™ä½Žç»´åº¦
def processDrops(df):
    # global df
    # rawDropList = ['Name', 'Names', 'Title', 'Sex', 'SibSp', 'Parch', 'Pclass', 'Embarked', \
    #                'Cabin', 'CabinLetter', 'CabinNumber', 'Age', 'Fare', 'Ticket', 'TicketNumber']
    #stringsDropList = ['Title', 'Name', 'Cabin', 'Ticket', 'Sex', 'Ticket', 'TicketNumber']
    # # if not keep_raw:
    # #     df.drop(rawDropList, axis=1, inplace=True)
    # # # elif not keep_strings:
    # #     df.drop(stringsDropList, axis=1, inplace=True)
    # df.drop(stringsDropList,axis=1,inplace=True)
    #ls = df.loc[:,[['PassengerId','Survived']]]
    # Ps = df.loc[:, ['PassengerId', 'Survived']]
    # Ps
    # Ps.corr()
    #åˆ é™¤å­—ç¬¦å±žæ€§çš„å€¼
    stringsDropList = ['Title', 'Name', 'Cabin', 'Ticket', 'Sex', 'Ticket', 'TicketNumber','Age_bin','Fare_bin']
    df.drop(stringsDropList,axis=1,inplace=True)

    #å‘çŽ°æŽ’é™¤survived å’ŒpassengerIdä»¥åŽ
    df_corr = df.drop(['Survived', 'PassengerId'], axis=1).corr(method='spearman')
    mask = np.ones(df_corr.columns.size) - np.eye(df_corr.columns.size)#é»˜è®¤ä¸­é—´ä¸º0ç›¸å½“äºŽè·Ÿè‡ªå·±æ— å…³
    df_corr = mask * df_corr#ä¹˜æ˜¯å¯¹åº”å…ƒç´ ä¹˜
    #å¾—åˆ°æ•´ä½“çš„åæ–¹å·®çŸ©é˜µ

    #èµ›é€‰å‡ºç›¸å…³æ€§å¤ªé«˜çš„å±žæ€§å†—æ‚å±žæ€§
    drops = []
    for col in df_corr.columns.values:
        if np.in1d([col],drops):#å·²ç»æœ‰äº†
            continue
        corr = df_corr[abs(df_corr[col]) > 0.98].index#ç¬¬colåˆ—ä¸­ç›¸å…³æ€§å¤ªå¤§çš„èˆå¼ƒ
        drops = np.union1d(drops, corr)
    print("\nDropping", drops.shape[0], "highly correlated features...\n")# , drops
    df.drop(drops,axis=1,inplace=True)
    return  df

#æˆ‘ä»¬å¾—åˆ°äº†å…·æœ‰å¤§é‡ç‰¹å¾çš„ç»´åº¦å¾ˆé«˜çš„æ•°æ®é›†ï¼Œç‰¹å¾è¾ƒå¤šä¸èƒ½ç›´æŽ¥ç”¨æ¥ä½œä¸ºæ¨¡åž‹è¾“å…¥ï¼Œä¸€æ˜¯å› ä¸ºè¿™äº›ç‰¹å¾é—´å…·æœ‰å¤šé‡å…±çº¿æ€§ï¼Œ
# å¯èƒ½ ä¼šå¯¼è‡´ç©ºé—´çš„ä¸ç¨³å®šï¼›äºŒæ˜¯å› ä¸ºé«˜ç»´ç©ºé—´æœ¬èº«å…·æœ‰ç¨€ç–æ€§ï¼Œä¸€ç»´æ­£æ€åˆ†å¸ƒæœ‰68%çš„å€¼è½äºŽæ­£è´Ÿæ ‡å‡†å·®ä¹‹é—´ï¼Œè€Œåœ¨åç»´ç©ºé—´ä¸Šåªæœ‰0.02%ï¼›ä¸‰æ˜¯ç”±äºŽè¿‡å¤šçš„å±žæ€§
# ä¼šä½¿æŒ–æŽ˜éœ€è¦å¾ˆé•¿æ—¶é—´ã€‚å¯¹äºŽä¸€äº›æ¨¡åž‹æ¥è¯´ï¼Œæ¯”å¦‚ä½¿ç”¨L1ï¼ˆLassoï¼‰ï¼Œå½“æœ‰å¤§é‡å±žæ€§æ—¶æ•ˆæžœå¾ˆå¥½ï¼Œå› ä¸ºå®ƒå¯ä»¥æœ‰æ•ˆå¿½ç•¥æŽ‰å™ªå£°å˜é‡ã€‚è€Œä¸€äº›æ¨¡åž‹åˆ™å®¹æ˜“è¿‡æ‹Ÿ åˆã€‚
#ä¼šä½¿ç”¨èšç±»åˆ†æžå’Œé™ç»´æ¥å¤„ç†æ•°æ®
#å½’çº¦æ­¥éª¤
def reduceAndCluster(df,input_df,clusters=3):
    #å±žæ€§è§„çº¦
    survivedSeries = pd.Series(df['Survived'],name='Survived')#å°†å€¼æ‹¿å‡ºæ¥
    df.drop('Survived',axis=1,inplace=True)
    #df = df.reindex_axis(input_df.columns,axis=1)#é‡æ–°æŒ‰ç…§input_dfåˆ—æ–¹å¼
    X = df.values
    #y = df.values[:, 0]#ç±»åˆ«å±žæ€§

    variance_pct = .99

    # Create PCA object
    pca = PCA(n_components=variance_pct)

    # Transform the initial features
    X_transformed = pca.fit_transform(X)

    # Create a data frame from the PCA'd data
    pcaDataFrame = pd.DataFrame(X_transformed)

    print("åŽŸæ•°æ®ç»´åº¦",X.shape[1])
    print("PCAåŽç»´åº¦",X_transformed.shape[1])


    #å€¼è§„çº¦ã€‚èšç±»
    from sklearn.cluster import KMeans

    #èšç±»å±žæ€§ æŽ¢æµ‹æ ·æœ¬é—´çš„ç›¸å…³æ€§
    kmeans = KMeans(n_clusters=clusters, random_state=np.random.RandomState(4), init='random')

    #æˆ‘ä»¬åˆ†ä¸ºè®­ç»ƒé›†å’Œæµ‹è¯•é›†
    #split_train, split_cv = train_test_split(df, test_size=0.2, random_state=0)

    trainClusterIds = kmeans.fit_predict(X_transformed[:input_df.shape[0]])#å¾—åˆ°æ¯ä¸ªæ ·æœ¬çš„èšç±»ä¸­å¿ƒ
    print("è®­ç»ƒæ ·æœ¬çš„èšç±»ä¸­å¿ƒæ˜¯:",trainClusterIds)

    testClusterIds = kmeans.predict(X_transformed[input_df.shape[0]:])
    # print
    # "clusterIds shape for test data: ", testClusterIds.shape
    print ("æµ‹è¯•æ ·æœ¬çš„èšç±»ä¸­å¿ƒæ˜¯: ", testClusterIds)

    clusterIds = np.concatenate([trainClusterIds,testClusterIds])
    print("æ•´ä½“çš„æ ·æœ¬ä¸­å¿ƒ: ", clusterIds.shape)

    # åˆ›å»ºèšç±»ä¸­å¿ƒçš„Id
    clusterIdSeries = pd.Series(clusterIds, name='ClusterId')
    df = pd.concat([survivedSeries, clusterIdSeries, pcaDataFrame], axis=1)

    return df



#æ•°æ®æ¸…ç†ä»¥åŠç‰¹å¾å·¥ç¨‹
def clean_and_feature_engineer(df,input_df):
    df = processCabin(df)
    df = processEmbarked(df)
    df = processTicket(df)
    df = processFare(df)
    df = processFamily(df)
    df = processName(df)
    df = processSex(df)
    df = processPclass(df)
    df = processAge(df)
    #ç»„åˆç‰¹å¾
    df = combineFeature(df)

    #åˆ é™¤å†—æ‚å±žæ€§
    df = processDrops(df)
    drop_list = ['PassengerId']
    df.drop(drop_list,axis=1,inplace=1)
    #input_df = df[:input_df.shape[0]]
    #submit_df = df[input_df.shape[0]:]

    #è§„çº¦
    df = reduceAndCluster(df,input_df)

    return df


# #ä¸»å‡½æ•°
def dataprocess(filename='train.csv',testname='test.csv'):
#     #å°†è®­ç»ƒå’Œæµ‹è¯•ç»„åˆ
#é‡æ–°ç´¢å¼•ä»¥åŠåŽ»æŽ‰NANçš„survivedæäº¤å˜é‡
    # å°†è®­ç»ƒå’Œæµ‹è¯•ç»„åˆ
    input_df = pd.read_csv(filename)
    submit_df = pd.read_csv(testname)
    df = pd.concat([input_df, submit_df])
    df.reset_index(inplace=True)
    df.drop('index', axis=1, inplace=True)

    df = clean_and_feature_engineer(df,input_df)
    input_df = df[:input_df.shape[0]]
    submit_df = df[input_df.shape[0]:]
    submit_df.reset_index(inplace=True)
    submit_df.drop('index', axis=1, inplace=True)
    submit_df.drop('Survived', axis=1, inplace=1)
    return input_df,submit_df
#     return data_train

```
